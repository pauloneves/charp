```python
import numpy as np

def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 3, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    ax[0].set_title("Meu ttulo")
    ax[0].set_xlabel("Valores de X ttulo")
    ax[0].set_ylabel("Y Values")

    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
    
    num_bars = 6
    bars = [4 + 10 * np.random.rand() for _ in range(num_bars)]
    ax[2].bar(np.arange(num_bars), bars)
    #ax[2].set_title("Barras codificadas")
                   
hist_and_lines()
```

```python
plt.style.use('./paulo.mplstyle')
hist_and_lines()
```

## Clean font cache and check installed fonts

```python

import shutil
import matplotlib

shutil.rmtree(matplotlib.get_cachedir())
```

```python
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```
