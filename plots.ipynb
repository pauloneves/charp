```python
import numpy as np

def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 3, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    ax[0].set_title("Meu ttulo")
    ax[0].set_xlabel("Valores de X ttulo")
    ax[0].set_ylabel("Y Values")
    ax[0].axhline(150)
    import string
    num_lines = 4
    for i in range(num_lines):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(string.ascii_letters[:num_lines], loc='lower left')
    
    num_bars = 6
    bars = [4 + 10 * np.random.rand() for _ in range(num_bars)]
    ax[2].bar(np.arange(num_bars), bars)
    #ax[2].set_title("Barras codificadas")
    
                   
hist_and_lines()
```

## Horizontal Bars

```python
def barh():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(15, 3))
    num_bars = 6
    bars = [20 * np.random.rand() for _ in range(num_bars)]
    ax[0].barh(np.arange(num_bars), bars)
    charp.set_title("Horizontal Bars", ax=ax[0])
    multi_words = ("ipsulum lorem " * 10).split()
    labels = [' '.join(multi_words[:i+1]).strip() for i in np.arange(num_bars)]
    ax[1].barh(y=np.arange(num_bars), width=bars)
    ax[1].set_yticks(np.arange(num_bars))
    ax[1].set_yticklabels(labels)
    charp.barh(break_lines=True, ax=ax[1])
    charp.set_title("Horizontal Bars Fixed", ax=ax[1])
    
plt.style.use('./paulo.mplstyle')    
barh()
```

## Vertical bars

```python
def bar():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(15, 3))
    num_bars = 6
    bars = [20 * np.random.rand() for _ in range(num_bars)]
    ax[0].bar(np.arange(num_bars), bars)
    ax[0].set_title("Vertical Bars")
    ax[1].bar(np.arange(num_bars), bars)
    charp.set_title("Vertical Bars Fixed", ax=ax[1])
    charp.bar(ax[1])
plt.style.use('./paulo.mplstyle')    
bar()
```

```python
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```
